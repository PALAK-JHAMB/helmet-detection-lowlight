```python
import torch, torch.nn as nn, torch.optim as optim
import cv2, numpy as np, glob, os
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image

class ZeroDCE(nn.Module):
    def __init__(self):
        super().__init__()
        self.relu = nn.ReLU(inplace=True)
        self.e_conv1 = nn.Conv2d(3,  32, 3, 1, 1)
        self.e_conv2 = nn.Conv2d(32, 32, 3, 1, 1)
        self.e_conv3 = nn.Conv2d(32, 32, 3, 1, 1)
        self.e_conv4 = nn.Conv2d(32, 32, 3, 1, 1)
        self.e_conv5 = nn.Conv2d(64, 32, 3, 1, 1)
        self.e_conv6 = nn.Conv2d(64, 32, 3, 1, 1)
        self.e_conv7 = nn.Conv2d(64, 24, 3, 1, 1)
    def forward(self, x):
        x1=self.relu(self.e_conv1(x)); x2=self.relu(self.e_conv2(x1))
        x3=self.relu(self.e_conv3(x2)); x4=self.relu(self.e_conv4(x3))
        x5=self.relu(self.e_conv5(torch.cat([x3,x4],1)))
        x6=self.relu(self.e_conv6(torch.cat([x2,x5],1)))
        x_r=torch.tanh(self.e_conv7(torch.cat([x1,x6],1)))
        enhanced=x
        for r in torch.split(x_r,3,dim=1):
            enhanced=enhanced+r*(torch.pow(enhanced,2)-enhanced)
        return enhanced, x_r

def make_dark(path, gamma=0.3):
    img=cv2.imread(path)
    dark=np.power(img/255.0,1/gamma)*255
    return np.clip(dark,0,255).astype(np.uint8)

DEVICE=torch.device("cuda" if torch.cuda.is_available() else "cpu")
src="/content/Hard-Hat-Workers-1/train/images"
out="/content/dark_imgs"; os.makedirs(out,exist_ok=True)
for i,p in enumerate(glob.glob(src+"/*.jpg")[:200]):
    cv2.imwrite(f"{out}/d{i:04d}.jpg", make_dark(p))
print(f"Created {len(os.listdir(out))} dark images")

class DarkDataset(Dataset):
    def __init__(self,d):
        self.imgs=glob.glob(d+"/*.jpg")
        self.t=transforms.Compose([transforms.Resize((256,256)),transforms.ToTensor()])
    def __len__(self): return len(self.imgs)
    def __getitem__(self,i): return self.t(Image.open(self.imgs[i]).convert("RGB"))

model_z=ZeroDCE().to(DEVICE)
opt=optim.Adam(model_z.parameters(),lr=1e-4)
loader=DataLoader(DarkDataset(out),batch_size=8,shuffle=True,num_workers=2)

for epoch in range(30):   # only 10 epochs
    total=0
    for batch in loader:
        img=batch.to(DEVICE); enh,A=model_z(img)
        pool=nn.AvgPool2d(16)(torch.mean(enh,1,keepdim=True))
        l_exp=torch.mean((pool-0.6)**2)
        mr,mg,mb=torch.split(torch.mean(enh,[2,3],keepdim=True),1,dim=1)
        l_col=torch.mean(((mr-mg)**2+(mr-mb)**2+(mb-mg)**2)**0.5)
        l_tv=torch.mean(torch.abs(A[:,:,1:,:]-A[:,:,:-1,:]))+torch.mean(torch.abs(A[:,:,:,1:]-A[:,:,:,:-1]))
        loss=10*l_exp+5*l_col+200*l_tv
        opt.zero_grad(); loss.backward(); opt.step()
        total+=loss.item()
    print(f"Epoch [{epoch+1}/30] Loss: {total/len(loader):.4f}")

torch.save(model_z.state_dict(),"/content/zerodce.pth")
print("Zero-DCE saved!")
```

```python
import matplotlib.pyplot as plt, tempfile
from torchvision import transforms
from ultralytics import YOLO

yolo=YOLO("/content/runs/helmet_pretrained/weights/best.pt")
model_z.eval()
T=transforms.ToTensor()
test_imgs=glob.glob("/content/Hard-Hat-Workers-1/test/images/*.jpg")[:4]

fig,axes=plt.subplots(len(test_imgs),3,figsize=(15,5*len(test_imgs)))
for i,path in enumerate(test_imgs):
    dark=make_dark(path,0.3)
    t_img=T(Image.open(path).convert("RGB")).unsqueeze(0).to(DEVICE)
    with torch.no_grad(): enh,_=model_z(t_img)
    enh=np.clip(enh.squeeze(0).cpu().permute(1,2,0).numpy()*255,0,255).astype(np.uint8)
    with tempfile.NamedTemporaryFile(suffix=".jpg",delete=False) as tmp:
        cv2.imwrite(tmp.name,cv2.cvtColor(enh,cv2.COLOR_RGB2BGR))
        det=yolo(tmp.name,conf=0.3,verbose=False)[0].plot()
    axes[i][0].imshow(cv2.cvtColor(dark,cv2.COLOR_BGR2RGB)); axes[i][0].set_title("Dark Input"); axes[i][0].axis("off")
    axes[i][1].imshow(enh); axes[i][1].set_title("Zero-DCE Enhanced"); axes[i][1].axis("off")
    axes[i][2].imshow(cv2.cvtColor(det,cv2.COLOR_BGR2RGB)); axes[i][2].set_title("Helmet Detected"); axes[i][2].axis("off")

plt.suptitle("Low-Visibility Enhancement → Helmet Detection", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("/content/results.png",dpi=150)
plt.show()
print("Saved!")
```
